```python
import sys
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import os
import struct
from matplotlib import animation as animation

%matplotlib inline
```

```python
class MNIST():
    def __init__(self, directory):
        self._directory = directory
        
        self._training_data = self._load_binaries("train-images.idx3-ubyte")
        self._training_labels = self._load_binaries("train-labels.idx1-ubyte")
        self._test_data = self._load_binaries("t10k-images.idx3-ubyte")
        self._test_labels = self._load_binaries("t10k-labels.idx1-ubyte")
        
        np.random.seed(0)
        samples_n = self._training_labels.shape[0]
        random_indices = np.random.choice(samples_n, samples_n // 10, replace = False)
        np.random.seed()
        
        self._validation_data = self._training_data[random_indices]
        self._validation_labels = self._training_labels[random_indices]
        self._training_data = np.delete(self._training_data, random_indices, axis = 0)
        self._training_labels = np.delete(self._training_labels, random_indices)
    
    def _load_binaries(self, file_name):
        path = os.path.join(self._directory, file_name)
        
        with open(path, 'rb') as fd:
            check, items_n = struct.unpack(">ii", fd.read(8))

            if "images" in file_name and check == 2051:
                height, width = struct.unpack(">II", fd.read(8))
                images = np.fromfile(fd, dtype = 'uint8')
                return np.reshape(images, (items_n, height, width))
            elif "labels" in file_name and check == 2049:
                return np.fromfile(fd, dtype = 'uint8')
            else:
                raise ValueError("Not a MNIST file: " + path)
    
    
    def get_training_batch(self, batch_size):
        return self._get_batch(self._training_data, self._training_labels, batch_size)
    
    def get_validation_batch(self, batch_size):
        return self._get_batch(self._validation_data, self._validation_labels, batch_size)
    
    def get_test_batch(self, batch_size):
        return self._get_batch(self._test_data, self._test_labels, batch_size)
    
    def _get_batch(self, data, labels, batch_size):
        samples_n = labels.shape[0]
        if batch_size <= 0:
            batch_size = samples_n
        
        random_indices = np.random.choice(samples_n, samples_n, replace = False)
        data = data[random_indices]
        labels = labels[random_indices]
        for i in range(samples_n // batch_size):
            on = i * batch_size
            off = on + batch_size
            return data[on:off], labels[on:off]
    
    
    def get_sizes(self):
        training_samples_n = self._training_labels.shape[0]
        validation_samples_n = self._validation_labels.shape[0]
        test_samples_n = self._test_labels.shape[0]
        return training_samples_n, validation_samples_n, test_samples_n
    
        
    
```

```python
mnist = MNIST('')

# Initialization of the first batches
test_batch = mnist.get_test_batch(500)
validation_batch = mnist.get_validation_batch(5)
training_batch = mnist.get_training_batch(5)

# Helping functions to change the format of the labels (we had some problems with the helper script)
def driveVector(vector):
    b = len(vector)
    label = np.zeros([b, 10])
    for k in range(b):
        l = training_batch[1][k]
        label[k][l-1] = 1
    return label

def driveVectorTest(vector):
    b = len(vector)
    label = np.zeros([b, 10])
    for k in range(b):
        l = test_batch[1][k]
        label[k][l-1] = 1
    return label

def driveVectorValidation(vector):
    b = len(vector)
    label = np.zeros([b, 10])
    for k in range(b):
        l = validation_batch[1][k]
        label[k][l-1] = 1
    return label



```

```python
# Tensorflow placeholders and Variables and computions for later
x = tf.placeholder(tf.float32, [None, 784])

bias = tf.Variable(tf.zeros([10]))

weights = tf.Variable(tf.random_normal([784, 10], stddev = 0.000002), dtype = tf.float32)

labels_soft = tf.placeholder(tf.float32, [None, 10])

y =tf.matmul(x, weights)+bias

softmax = tf.nn.softmax_cross_entropy_with_logits(labels = labels_soft, 
                                                  logits = y)
                                                 
train_step = tf.train.GradientDescentOptimizer(0.5).minimize(softmax)

entropy_mean = tf.reduce_mean(softmax)

correct_prediction = tf.equal(tf.argmax(y, 1), tf.argmax(labels_soft, 1))

accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))

#Training parameters
b_size = 1100
training_steps = 100
n = 10
epochs = 3

# storages for plotting
training_accuracy = []
validation_accuracy = []
cross_entropy_train = []
cross_entropy_val = []


with tf.Session() as session: 
    session.run(tf.global_variables_initializer())
    v_acc = 0
    v_ce = 0
    
    # We train in 3 epochs
    for j in range(epochs):
                
        # Here we actually train and change weights and bias
        # We also take the data into list for later plotting 
        
        for z in range (training_steps):
            
            loss, _ = session.run([entropy_mean, train_step], 
                              feed_dict = {x: np.reshape(training_batch[0],(len(training_batch[0]), 784)), 
                                    labels_soft: driveVector(training_batch[1])})
            
            cross_entropy_train.append(loss)

            training_accuracy.append(session.run(accuracy, 
                                                 feed_dict={x: np.reshape(training_batch[0],(len(training_batch[0]), 784)),
                                                            labels_soft: driveVector(training_batch[1])}))
            validation_accuracy.append(v_acc)
            cross_entropy_val.append(v_ce)
            training_batch = mnist.get_training_batch(b_size)
            
            # Validation after n training steps without chaning parameters
            if(z%n==0): 
                v_acc = session.run(accuracy, feed_dict={x: np.reshape(validation_batch[0],(len(validation_batch[0]), 784)), labels_soft: driveVectorValidation(validation_batch[1])})
                validation_batch = mnist.get_validation_batch(100)
                v_ce = session.run(entropy_mean, 
                              feed_dict = {x: np.reshape(validation_batch[0],(len(validation_batch[0]), 784)), 
                                    labels_soft: driveVector(validation_batch[1])})

    # The accuracy of the trained model
    print('Accuracy over Test data:')
    print(session.run(accuracy, feed_dict={x: np.reshape(test_batch[0],(len(test_batch[0]), 784)), labels_soft: driveVectorTest(test_batch[1])}))
```

```python
# Plotting

training_step_sizes = np.arange(epochs*training_steps)

# Accuracy plot
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(121)
ax.set_title("Accuracy")
ax.plot(training_step_sizes,training_accuracy,label="Training Data")
ax.plot(training_step_sizes,validation_accuracy,label="Validation Data")
ax.set_xlabel("Training steps")
ax.set_ylabel("Accuracy")
ax.legend()

# Entropy plot
ax = fig.add_subplot(122)
ax.set_title("Cross entropy")
ax.plot(training_step_sizes,cross_entropy_train,label="Training Data")
ax.plot(training_step_sizes, cross_entropy_val ,label="Validation Data")
ax.set_xlabel("Training steps")
ax.set_ylabel("Cross entropy")
ax.legend()

fig.tight_layout()

# Plotting of the weights did not work for us, since we could not get the weights for certain labels with the helper script
```
